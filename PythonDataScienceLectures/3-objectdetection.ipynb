```python
import numpy as np
import matplotlib.pyplot as pl
import matplotlib.patches as patches
```

```python
import cv2
```

```python
from PIL import Image
```

```python
#imagefile="cat1.jpeg"
#imagefile="manycats.jpg"
imagefile="cat_dog.jpg"
```

```python
cascfile="haarcascade_frontalcatface_extended.xml"
```

```python
catCascade=cv2.CascadeClassifier(cascfile)
```

```python
# read the imagefile into OpenCV:
image=cv2.imread(imagefile)
```

```python
# convert the RGB image to a greyscale image:
image_grey = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
cats = catCascade.detectMultiScale(
    image_grey,
    scaleFactor=1.005,
    minNeighbors=10,
    minSize=(75, 75)
)
```

```python
print "Found ",len(cats)," cats!"
```

```python
# create a PIL Image:
im = np.array(Image.open(imagefile), dtype=np.uint8)
 
# create a figure and axes:
fig,ax = pl.subplots(1)
 
# display the image:
ax.imshow(im)
 
# extract the position (x,y) and the width & height
# of the detection:
for (x, y, w, h) in cats:
 
    # make a rectangular patch:
    rect = patches.Rectangle((x,y),w,h,linewidth=1,edgecolor='r',facecolor='none')
 
    # add the patch to your axes:
    ax.add_patch(rect)
 
# save the annotated image:
pl.savefig("detection1.png")
pl.axis('off')

# display the annotated image:
pl.show()
```
